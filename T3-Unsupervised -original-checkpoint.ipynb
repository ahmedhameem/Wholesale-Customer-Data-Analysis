```python
a = ['Channel', 'Region', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# We set limits to the variables and try to filter the dataframe to adjust for outliers
q1 = df.quantile(0.25)  # Obtaining 1st quartile of each variable
q3 = df.quantile(0.75)  # Obtaining 3rd quartile of each variable
iqr = q3 - q1  # Finding the Interquartile range for each variable             
ul = q3+1.5*iqr  # Finding the upper fence
ll = q1-1.5*iqr  # Finding the lower fence

for column in a:
    df = df[(df[column] < ul[column]) & (df[column] > ll[column])]
```

```python
#Principal Component Analysis (PCA) is used to reduce the dimensionality of data while preserving as much variance as possible. 
#It helps in simplifying the dataset by transforming it into a new set of variables (principal components) that are uncorrelated and ordered by the amount of variance they explain
from sklearn.decomposition import PCA   # importing the required library
PC=PCA()
PC.fit(df_scaled)
```

```python
# we would need to determine the ideal number of principal components to use for our analysis
# in order to do so a cumulative graph for explained variance will be used 
cumulative_variance = []
total_variance = sum(PC.explained_variance_ratio_)
for i in range(len(df.columns)):
    cumulative_variance.append(sum(PC.explained_variance_ratio_[:i+1]))

plt.plot(range(1, len(df.columns)+1), cumulative_variance, marker='o', linestyle='--')
plt.title("Cumulative Explained variance (PCA)")
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.show()


# the marginal increase for explained variance drops at 5 principal components hence we will choose 5 principal components
# therefore we will choose 5 components for our analysis
```

```python
PC=PCA(n_components=5)   #the PCA model will retain the top 5 principal components, effectively reducing the dimensionality of the data to 5 dimensions

PC.fit(df_scaled)  #This process transforms the data into a new feature space defined by the top 5 principal components
                            # capturing the maximum variance in the original dataset while reducing its dimensionality.

PC_T=PC.transform(df_scaled)  #transforming the original dataset using Principal Component Analysis.
```

```python
# The biplot function extracts principal component scores and scales them for visualization in a scatter plot. 
def biplot(score, coef, labels=None):
# Extracting the principal component scores for P1 and P2
    X = score[:,0]
    Y = score[:,1]
# Calculating scaling factors for X and Y axes
    X_scaled = 1.0/(X.max() - X.min())
    Y_scaled = 1.0/(Y.max() - Y.min())
# Plotting the principal component scores, scaling them for consistent visualization
    plt.scatter(X * X_scaled, Y * Y_scaled, s=5, color='blue')
# Returning the scaled scores for further use
    return X_scaled, Y_scaled

## The add_loadings function plots the loading vectors of principal components onto a PCA biplot and adds labels to them.
def add_loadings(X_scaled, Y_scaled, coef, labels):
    n = len(labels)  # Number of components
# Looping over each principal component to plot the loading vector
    for i in range(n):
# Plotting the loading vector as an arrow
        plt.arrow(0, 0, coef[i,0], coef[i,1], color='red', alpha=0.5)
# Adding labels to the loading vectors
        plt.text(coef[i,0]*1.15, coef[i,1]*1.15, labels[i], color='black', ha='center', va='center')
    
# Setting labels and title for the plot
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title('PCA Biplot')
```

```python
from sklearn.cluster import KMeans# importing the required library
# the range of clusters to explore
num_clusters = range(1, 9)

# empty list to store the within-cluster sum of squares (WCSS)
wcss = []

# Iterate over each number of clusters
for i in num_clusters:
# create a KMeans instance with the current number of clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', 
                    max_iter=300,
                    n_init=10)
    
# fit the KMeans model to the data
    kmeans.fit(PC_df)
    
# append the WCSS to the list
    wcss.append(kmeans.inertia_)

# elbow graph
plt.plot(num_clusters, wcss, marker='o')
plt.title('The Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(num_clusters)
plt.show()  
#elbow graph helps in selecting an appropriate number of clusters for K-means clustering by identifying the point where adding more clusters ceases to provide significant improvement in clustering quality
# from the graph below we see a bend at 2 clusters and it seems to be an elbow point
# Therefore 2 clusters will be used for visualization purposes using Kmeans
```

```python
# define KMeans parameters
kmeans_params = {
    'n_clusters': 2,       # number of clusters to form
    'init': 'k-means++',   # method for initialization of centroids
    'max_iter': 300,       # maximum number of iterations for each run
    'n_init': 10,          # number of times KMeans will be run with different initial centroids
    'random_state': 12     
}

# create KMeans object with specified parameters
kmeans = KMeans(**kmeans_params)

# fit KMeans to the dataset and predict cluster idss
pred_kmeans = kmeans.fit_predict(PC_df)
```

```python
#import the required library
import collections
print('AgglomerativeClustering w/ Ward Result :',collections.Counter(model_predict))

#summary of how many data points were assigned to each cluster by the agglomerativeClustering model with Ward linkage. 
#this can be useful for understanding the distribution of data points among the clusters and evaluating the performance of the clustering algorithm
```

```python
#import the required libraries
from sklearn.metrics import silhouette_samples
from  matplotlib import cm

# getting unique cluster labels
cluster_labels = np.unique(model_predict)
# counting the number of clusters
n_clusters = cluster_labels.shape[0]

# calculating silhouette scores for each data point
silhouette_vals = silhouette_samples(PC_T, model_predict, metric='euclidean')

# initializing y-axis lower and upper bounds
y_ax_lower, y_ax_upper = 0, 0
yticks = []  # Initializing y-axis tick labels list

# iterating over each cluster label
for i, c in enumerate(cluster_labels):
    # Selecting silhouette scores for data points in the current cluster
    c_silhouette_vals = silhouette_vals[model_predict == c]
    c_silhouette_vals.sort()# sorting silhouette scores
    y_ax_upper += len(c_silhouette_vals) # updating y-axis upper bound
    
# choosing color for the current cluster
    color = cm.jet(float(i) / n_clusters)
    
# plotting horizontal bar chart for silhouette scores of the current cluster
    plt.barh(range(y_ax_lower, y_ax_upper),
             c_silhouette_vals,
             height=1.0,
             edgecolor='none',
             color=color)
    
   
    yticks.append(y_ax_lower + y_ax_upper / 2.)  # Adding y-axis tick label at the middle of the current cluster's silhouette values
    
   
    y_ax_lower += len(c_silhouette_vals) # Updating y-axis lower bound

# Calculating average silhouette score
silhouette_avg = np.mean(silhouette_vals)



plt.axvline(silhouette_avg, color='red', linestyle='--') # adding dotted vertical line for average silhouette score
plt.ylabel("Cluster")
plt.xlabel('Silhouette Coefficient')
plt.show()
```
